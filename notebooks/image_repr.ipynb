```python
import sys
sys.path.append('..')

from IPython.display import display

from aural_travels.model.image_repr import DALLEImageRepr, VQGANImageRepr
from aural_travels.data import soundcloud
```

```python
DATA_DIR = '../../scdata'
track_ids = [432611943,
             112715018,
             620041359,
             727356370,
             933592519,
             145658079,
             104096544,
             319367841,
             460737276,
             313450061]
images = [soundcloud.load_image(DATA_DIR, track_id) for track_id in track_ids]
```

```python
image_reprs = [('DALL-E', DALLEImageRepr()),
               ('VQGAN', VQGANImageRepr())]
```

```python
for image in images:
    print('input')
    display(image)
    
    for image_repr_name, image_repr in image_reprs:
        tensor = image_repr.image_to_tensor(image)
        image_seq = image_repr.encode(tensor)
        tensor = image_repr.decode(image_seq)[0]
        
        print(image_repr_name)
        print(image_seq.shape)
        display(image_repr.tensor_to_image(tensor))
```
